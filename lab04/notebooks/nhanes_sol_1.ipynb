```sql
-- Projete reversamente um modelo ER das tabelas
medications-drug [__Codigo__, NomeMedicamento]
CHP(Codigo)

demographic-person [__Id__, Genero, Idade]
CHP(Id)

medications-use [IdPessoa, CodMedicamento, DiasUso]
CHE(IdPessoa) demographic-person -> medications-use
CHE(CodMedicamento) medications-drug -> medications-use
```

```sql
-- Escreva instruções SQL para montar as tabelas conforme definido no esquema relacional
-- Realize a importação de CSV
-- PS: não defina chave primária para a tabela de uso de medicamento
-- Crie três tabelas uma para cada CSV
-- PS: não defina chave primária para a tabela de uso de medicamento
CREATE TABLE medications-drug(
    Codigo varchar(6) NOT NULL,
    NomeMedicamento varchar(50) NOT NULL,
    PRIMARY KEY(Codigo)    
);

CREATE TABLE demographic-person(
    Id int NOT NULL,
    Genero int NOT NULL,
    Idade int NOT NULL,
    PRIMARY KEY(Id)    
);

CREATE TABLE medications_use(
    IdPessoa int NOT NULL,
    CodMedicamento varchar(6) NOT NULL,
    DiasUso int NOT NULL,
    PRIMARY KEY(IdPessoa, CodMedicamento),
    FOREIGN KEY(IdPessoa)
        REFERENCES demographic-person(Id),
    FOREIGN KEY(CodMedicamento)
        REFERENCES medications-drug(Codigo)
);


medications-drug [Codigo, NomeMedicamento]
CHP(Codigo)
```

```sql
-- Liste todos os medicamentos
SELECT DISTINCT NomeMedicamento FROM medications-drug;
```

```sql
-- Liste todas as pessoas
SELECT Id FROM demographic-person;
```

```sql
-- Liste todos os usos de medicamento
SELECT IdPessoa,CodMedicamento FROM medications-use;
```

```sql
-- Liste pessoas com idade acima de 60
SELECT Id FROM demographic-person WHERE Idade > 60;
```

```sql
-- Liste as idades das pessoas cadastradas (sem repetições)
SELECT DISTINCT Idade FROM demographic-person;
```

```sql
-- Liste código das pessoas, idade, código dos medicamentos que usaram e dias de uso
SELECT Pe.Id, Pe.Idade, Me.Codigo, Mu.DiasUso FROM demographic-person Pe, medications_drug Me, medications-use Mu 
WHERE Pe.Id = Mu.IdPessoa AND Me.Codigo = Mu.CodMedicamento;
```

```sql
-- Liste código das Pessoas, idade, nome dos medicamentos que usaram e quantos dias de uso
SELECT Pe.Id, Pe.Idade, Me.NomeMedicamento, Mu.DiasUso FROM demographic-person Pe, medications_drug Me, medications-use Mu 
WHERE Pe.Id = Mu.IdPessoa AND Me.Codigo = Mu.CodMedicamento;
```

```sql
-- Liste nome dos medicamentos que foram usados por pessoas (sem repetições)
SELECT Me.NomeMedicamento, COUNT(*) NUM FROM demographic-person Pe, medications_drug Me, medications-use Mu 
WHERE Pe.Id = Mu.IdPessoa AND Me.Codigo = Mu.CodMedicamento GROUP BY Me.NomeMedicamento;
```

```sql
-- Liste nome dos medicamentos usados por mais de 2.000 dias (sem repetições)
SELECT Me.NomeMedicamento, COUNT(*) NUM FROM demographic-person Pe, medications_drug Me, medications-use Mu 
WHERE Pe.Id = Mu.IdPessoa AND Me.Codigo = Mu.CodMedicamento AND Mu.DiasUso > 2000 GROUP BY Me.NomeMedicamento;
```
