```python
import io
import requests
import xml.etree.ElementTree as et
url = "https://apps.who.int/gho/athena/api/GHO/WHOSIS_000001/AFG"
data = requests.get(url).content
x = et.parse(io.StringIO(data.decode("utf-8")))

print("Assim como no exemplo, o xml achou um elemento na URL específica: ")
print(x)
```
